This notebook is designed to help understand and debug the input fetched for bulk stocks.
You can use this to inspect the data and identify any issues.

```python
# Import necessary modules
import asyncio
import nest_asyncio
import sys

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()

sys.path.append("/Users/varhag/workspace/equisense-ai")  # Adjust the path as needed
sys.path.append("/Users/varhag/workspace/equisense-ai/agentic-stock-research")  # Adjust the path as needed
from app.tools.realtime_data import get_realtime_provider

# Define the tickers for bulk stocks
tickers = ["ITC.NS", "RELIANCE.NS", "TCS.NS", "INFY.NS"]

# Fetch real-time prices for bulk stocks
async def fetch_bulk_stock_data():
    provider = await get_realtime_provider()
    results = {}
    for ticker in tickers:
        price_data = await provider.get_real_time_price(ticker)
        results[ticker] = price_data
    return results

# Run the async function
bulk_stock_data = asyncio.run(fetch_bulk_stock_data())

# Display the fetched data
bulk_stock_data
```

```python
# Display the fetched data
for ticker, data in bulk_stock_data.items():
    print(f"Ticker: {ticker}")
    print(f"Price: {data.price}")

    print("---")
```
